## 실험 1: SKILL.md 프론트매터 작성 (5분)

Skills의 **Progressive Disclosure** 원칙에서 가장 먼저 로드되는 것이 프론트매터입니다.
Agent는 시작 시 프론트매터의 `name`과 `description`만 읽어 어떤 Skill이 있는지 파악합니다.

아래 빈칸을 채워 메일 체크 Skill의 프론트매터를 완성하세요.

**Agent Skills 스펙 규칙:**
- `name`: 64자 이내, 소문자+숫자+하이픈만 허용
- `description`: 1024자 이내, Skill이 하는 일 + 매칭 키워드 포함
- `metadata.tags`: 검색/분류에 사용되는 태그 목록

````python
# ===== 빈칸을 채우세요 =====

# SKILL.md의 YAML 프론트매터 구조
skill_frontmatter = {
    "name": ___,            # 힌트: 소문자+하이픈으로 된 Skill 이름 (예: "mail-check")
    "description": ___,     # 힌트: "메일함을 확인하고 중요 메일을 필터링하여 요약합니다." 같은 설명
    "metadata": {
        "author": "workshop",
        "version": "1.0",
        "tags": [___, ___],  # 힌트: 이 Skill과 관련된 태그 2개 (예: "email", "monitoring")
    },
}

# 검증: Progressive Disclosure의 Discovery 단계에서 사용되는 필드
assert isinstance(skill_frontmatter["name"], str), "name은 문자열이어야 합니다"
assert len(skill_frontmatter["name"]) <= 64, "name은 64자 이내여야 합니다"
assert skill_frontmatter["name"] == skill_frontmatter["name"].lower(), "name은 소문자만 허용됩니다"
assert "-" in skill_frontmatter["name"], "name에 하이픈이 포함되어야 합니다 (예: mail-check)"
assert len(skill_frontmatter["description"]) > 10, "description이 너무 짧습니다"
assert len(skill_frontmatter["metadata"]["tags"]) == 2, "tags는 2개를 채워주세요"

print(f"Skill 이름: {skill_frontmatter['name']}")
print(f"설명: {skill_frontmatter['description'][:60]}...")
print(f"태그: {skill_frontmatter['metadata']['tags']}")
print()
print("[Progressive Disclosure] Discovery 단계에서 Agent가 보는 정보:")
print(f"  <skill name='{skill_frontmatter['name']}'>")
print(f"    {skill_frontmatter['description'][:50]}...")
print(f"  </skill>")
print()
print("실험 1 통과: SKILL.md 프론트매터 완성")
````

## 실험 2: FastMCP Tool 등록 (5분)

MCP 서버는 `@mcp.tool()` 데코레이터로 Tool을 등록합니다.
이때 **함수 이름**이 Tool 이름이 되고, **docstring**이 Tool 설명이 됩니다.
Agent(LLM)는 이 설명을 보고 어떤 Tool을 호출할지 결정합니다.

아래 빈칸을 채워 `check_inbox` MCP Tool을 완성하세요.

````python
# ===== 빈칸을 채우세요 =====

# FastMCP Tool 등록 구조 (실제로는 @mcp.tool() 데코레이터 사용)
# 여기서는 구조만 확인합니다.

tool_definition = {
    "name": ___,         # 힌트: 메일함을 확인하는 함수 이름 ("check_inbox")
    "description": ___,  # 힌트: 이 Tool이 하는 일을 설명하는 docstring
    "parameters": {
        "filter": {
            "type": ___,         # 힌트: filter 파라미터의 타입 ("string")
            "description": "메일 필터 조건 (unread, important, all)",
            "default": "unread",
        }
    },
}

# 검증
assert tool_definition["name"] == "check_inbox", f"Tool 이름이 다릅니다: {tool_definition['name']}"
assert len(tool_definition["description"]) > 5, "description이 너무 짧습니다"
assert tool_definition["parameters"]["filter"]["type"] == "string", "filter의 타입은 'string'입니다"

print("FastMCP에서의 실제 코드:")
print()
print('  @mcp.tool()')
print(f'  async def {tool_definition["name"]}(filter: str = "unread") -> str:')
print(f'      \"\"\"{tool_definition["description"]}\"\"\"')
print('      ...')
print()
print("[핵심] 함수 이름 → Tool 이름, docstring → Tool 설명, 타입 힌트 → 파라미터 스키마")
print("       Agent(LLM)는 description을 보고 호출 여부를 결정합니다.")
print()
print("실험 2 통과: FastMCP Tool 구조 이해 완료")
````

## 실험 3: Skill → 시스템 프롬프트 주입 (5분)

MCP Tool은 **bind_tools**에 바인딩되어 실행 가능한 기능이 됩니다.

## 핵심 비교

| 구분 | Skills | MCP Tools |
|------|--------|----------|
| **역할** | "어떻게" — 절차적 지식 | "무엇에" — 데이터 접근 |
| **형태** | SKILL.md (마크다운 파일) | MCP 서버 (JSON-RPC) |
| **주입 위치** | 시스템 프롬프트 (`<skill>...</skill>`) | bind_tools (Tool 바인딩) |
| **로딩** | Progressive Disclosure (3단계) | Tool 스키마 전체 로드 |
| **비유** | 업무 매뉴얼 | 데이터 창구 |

## 학습 포인트

1. **SKILL.md 프론트매터**의 `name`과 `description`은 Progressive Disclosure의 Discovery 단계에서 사용됩니다.
2. **FastMCP**는 `@mcp.tool()` + 함수 이름 + docstring + 타입 힌트로 Tool을 자동 등록합니다.
3. **Skill → system_prompt**, **MCP Tool → bind_tools**로 각각 주입됩니다.
4. Skills와 MCP는 **독립적**입니다: Skill을 교체해도 MCP 서버 수정이 불필요하고, 그 반대도 마찬가지입니다.

- `type: "string"`: Python의 `str` 타입 힌트가 JSON Schema의 `"string"`으로 자동 변환됩니다.
