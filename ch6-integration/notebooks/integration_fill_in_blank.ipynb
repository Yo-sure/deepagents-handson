## 실험 1: End-to-End 시퀀스 정렬 (5분)

메일 Agent의 전체 동작 시퀀스를 올바른 순서로 정렬하세요.
각 단계에서 사용하는 **기술 계층**이 무엇인지 채워 넣습니다.

**사용 가능한 기술:** Skills, MCP, A2A, DeepAgents(Harness)

````python
# ===== 빈칸을 채우세요 =====

# 메일 Agent E2E 시퀀스 — 각 단계에서 사용하는 기술 계층
e2e_sequence = [
    {"step": 1, "action": "SKILL.md 로드 → 시스템 프롬프트 주입",  "tech": ___},     # 힌트: 절차적 지식 계층
    {"step": 2, "action": "check_inbox() 호출 → 메일 6통 반환",    "tech": ___},     # 힌트: Agent→Tool 수직 연결
    {"step": 3, "action": "중요 메일 3통 필터 + Notify Agent 알림", "tech": ___},     # 힌트: Agent→Agent 수평 통신
]

# 검증
assert e2e_sequence[0]["tech"] == "Skills", "SKILL.md 로드는 Skills 계층입니다"
assert e2e_sequence[1]["tech"] == "MCP", "check_inbox()는 MCP Tool 호출입니다"
assert e2e_sequence[2]["tech"] == "A2A", "Notify Agent 알림은 A2A 통신입니다"

print("메일 Agent E2E 시퀀스:")
for s in e2e_sequence:
    print(f"  Step {s['step']}: [{s['tech']:10s}] {s['action']}")

print("\n실험 1 통과: E2E 시퀀스 정렬 완료")
````

## 실험 2: 기술 스택 매핑 (5분)

Ch2~Ch5에서 배운 기술이 각각 **어떤 질문**에 답하는지 매핑하세요.

| 기술 | 질문 |
|------|------|
| LangGraph | "어떻게 **실행 흐름**을 구성하는가?" |
| DeepAgents | "공통 **보일러플레이트**를 어떻게 제거하는가?" |
| Skills | "작업을 **어떤 절차**로 수행하는가?" |
| MCP | "**어떤 도구/데이터**에 접근하는가?" |
| A2A | "다른 Agent에게 **어떻게 부탁**하는가?" |

````python
# ===== 빈칸을 채우세요 =====

# 각 기술이 답하는 핵심 질문
tech_stack = {
    "LangGraph":   "실행 흐름 구성 (StateGraph, 노드, 엣지)",
    "DeepAgents":  "Harness로 보일러플레이트 제거 (create_deep_agent)",
    "Skills":      ___,  # 힌트: "절차적 지식" — SKILL.md로 "어떻게" 수행할지 안내
    "MCP":         ___,  # 힌트: "데이터 접근" — MCP 서버로 "무엇에" 접근할지 표준화
    "A2A":         ___,  # 힌트: "Agent 간 협업" — Agent Card + Task로 역할 분리
}

# 검증
for key in ["Skills", "MCP", "A2A"]:
    assert isinstance(tech_stack[key], str) and len(tech_stack[key]) > 5, f"{key}의 설명을 채워주세요"

print("기술 스택 매핑:")
print("-" * 50)
for tech, role in tech_stack.items():
    print(f"  {tech:12s} → {role}")

print("\n실험 2 통과: 기술 스택 매핑 완료")
````

## 실험 3: HITL 지점 설계 (5분)

메일 Agent가 사용하는 Tool 중 **사람 승인이 필요한 Tool**은 무엇일까요?

**판단 기준:**
- 읽기 전용(check, search) → 자동 실행 OK
- 부수 효과(전달, 삭제, 알림) → 사람 승인 필요

아래 `interrupt_on` 파라미터를 구성하세요.

````python
# ===== 빈칸을 채우세요 =====

# 메일 Agent의 Tool 목록과 HITL 필요 여부
tools_hitl = {
    "check_inbox":      False,   # 읽기 전용 → 자동 실행
    "get_email_detail": False,   # 읽기 전용 → 자동 실행
    "search_emails":    False,   # 읽기 전용 → 자동 실행
    "forward_email":    ___,     # 힌트: 메일 전달은 부수 효과가 있을까요? (True/False)
    "send_notification": ___,    # 힌트: 알림 전송은 부수 효과가 있을까요? (True/False)
}

# DeepAgents interrupt_on 파라미터 구성
interrupt_on = {name: True for name, needs_hitl in tools_hitl.items() if needs_hitl}

# 검증
assert tools_hitl["forward_email"] == True, "메일 전달은 부수 효과(외부 전송)가 있으므로 승인이 필요합니다"
assert tools_hitl["send_notification"] == True, "알림 전송도 부수 효과(외부 알림)가 있으므로 승인이 필요합니다"
assert "forward_email" in interrupt_on
assert "send_notification" in interrupt_on
assert "check_inbox" not in interrupt_on

print("HITL 설계 결과:")
for name, needs_hitl in tools_hitl.items():
    status = "사람 승인 필요" if needs_hitl else "자동 실행"
    print(f"  {name:20s} → {status}")

print(f"\ninterrupt_on 파라미터: {interrupt_on}")
print("\n[핵심] 읽기 전용 Tool은 자동, 부수 효과 Tool은 HITL로 제어합니다.")
print("\n실험 3 통과: HITL 지점 설계 완료")
````

````python
# 전체 실험 결과 정리
print("=" * 50)
print("전체 실험 결과")
print("=" * 50)

print("\n[실험 1] E2E 시퀀스")
print("  Skills → MCP → A2A 순서로 기술 계층이 연결됨")

print("\n[실험 2] 기술 스택 매핑")
print("  LangGraph(실행) → DeepAgents(Harness) → Skills(절차) → MCP(데이터) → A2A(협업)")

print("\n[실험 3] HITL 설계")
print("  읽기 전용 = 자동, 부수 효과 = 사람 승인")
print("  interrupt_on으로 Tool 단위 제어")

print("\n모든 실험 통과!")
````

## 전체 아키텍처 정리

```
Ch2: LangGraph      → 실행 흐름 (StateGraph, 노드, 엣지)
Ch3: DeepAgents      → Harness (create_deep_agent, 미들웨어)
Ch4: Skills + MCP    → 절차(SKILL.md) + 데이터(@mcp.tool)
Ch5: A2A             → Agent 간 협업 (Agent Card, Task)
Ch6: 통합            → 모든 계층을 E2E로 연결
```

| 질문 | 기술 | 비유 |
|------|------|------|
| 어떻게 실행? | LangGraph | 공장 라인 설계 |
| 공통 인프라? | DeepAgents | 공장 자동화 설비 |
| 어떤 절차? | Skills | 작업 매뉴얼 |
| 어떤 데이터? | MCP | 데이터 창구 |
| 누구와 협업? | A2A | 부서 간 업무 요청 |
| 사람 개입? | HITL | 결재 라인 |

## 학습 포인트

1. 메일 Agent의 E2E 흐름은 **Skills(절차) → MCP(데이터) → A2A(협업)** 순서로 기술 계층이 연결됩니다.
2. Ch2~Ch5의 기술은 각각 다른 질문에 답합니다: 실행 흐름, 보일러플레이트, 절차, 데이터, 협업.
3. HITL은 **부수 효과가 있는 Tool**에 적용합니다: `interrupt_on` 파라미터로 Tool 단위 제어.
4. 이 모든 기술을 조합하면 **사람이 감독하는 자율 메일 Agent**가 됩니다.
